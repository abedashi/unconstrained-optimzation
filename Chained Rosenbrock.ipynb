#### Dependencies

```python
import numpy as np
from algorithms.utils.chained_rosenbrock import chained_rosenbrock
from algorithms.utils.plot import plot_combined_results, plot
from algorithms.utils.print_result import print_result
from tests.test_optimization_method import run_optimization_method

np.random.seed(336648)
```

#### Initialization

```python
# Define the problem base and scale
base_number = 10  # Base number for scaling
dimension_power = 3  # Exponent for the problem size
problem_size = base_number ** dimension_power  # Total problem size

# Initialize the starting point based on modulus conditions
initial_point = np.zeros(problem_size)
for i in range(problem_size):
    if (i % 2) == 1:
        initial_point[i] = -1.2
    else:
        initial_point[i] = 1.0

# Generate 10 random starting points in the hypercube [x - 1, x + 1]
random_points = [initial_point + np.random.uniform(-1, 1, problem_size) for _ in range(10)]
```

#### Initial Points

```python
print("\nCOST OF CHAINED ROSENBROCK BEFORE OPTIMIZATION:", chained_rosenbrock(initial_point))

print("\nRunning Modified Newton Test...")
result_n, trajectory_n = run_optimization_method("Modified Newton", "Chained Rosenbrock", initial_point, tol=1e-4, max_iter=1000)
print_result(result_n)


print("\nRunning Steepest Descent Test...")
result_s, trajectory_s = run_optimization_method("Steepest Descent", "Chained Rosenbrock", initial_point, tol=1e-4, max_iter=1000)
print_result(result_s)

plot_combined_results(chained_rosenbrock, trajectory_n, trajectory_s, "Chained Rosenbrock", "initial", [0, 0], [1, 1])
```

#### Random Points

```python
# Optimize for each random starting point
for idx, random_point in enumerate(random_points):
    print(f"\nCOST OF CHAINED ROSENBROCK BEFORE OPTIMIZATION (Random Point {idx+1}):",
          chained_rosenbrock(random_point))
    
    print(f"\nRunning Modified Newton Test for Random Point {idx+1}...")
    result_n, trajectory_n = run_optimization_method("Modified Newton", "Chained Rosenbrock", random_point, tol=1e-4, max_iter=1000)
    print_result(result_n)

    print(f"\nRunning Steepest Descent Test for Random Point {idx+1}...")
    result_s, trajectory_s = run_optimization_method("Steepest Descent", "Chained Rosenbrock", random_point, tol=1e-4, max_iter=1000)
    print_result(result_s)

    plot_combined_results(chained_rosenbrock, trajectory_n, trajectory_s, "Chained Rosenbrock", f"random{idx+1}")
```
