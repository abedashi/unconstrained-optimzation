#### Dependencies

```python
import numpy as np
from algorithms.utils.extended_powell_singular import extended_powell_singular
from algorithms.utils.plot import plot_combined_results, plot
from algorithms.utils.print_result import print_result
from tests.test_optimization_method import run_optimization_method

np.random.seed(336648)
```

#### Initialization

```python
# Define the problem base and scale
base_number = 10  # Base number for scaling
dimension_power = 3  # Exponent for the problem size
problem_size = base_number ** dimension_power  # Total problem size

# Initialize the starting point based on modulus conditions
initial_point = np.zeros(problem_size)
for i in range(problem_size):
    if (i % 4) == 1:
        initial_point[i] = 3
    elif (i % 4) == 2:
        initial_point[i] = -1
    elif (i % 4) == 3:
        initial_point[i] = 0
    else:
        initial_point[i] = 1

# Generate 10 random starting points in the hypercube [x - 1, x + 1]
random_points = [initial_point + np.random.uniform(-1, 1, problem_size) for _ in range(10)]
```

#### Initial Points

```python
print("\nCOST OF EXTENDED POWELL SINGULAR BEFORE OPTIMIZATION:", extended_powell_singular(initial_point))

print("\nRunning Modified Newton Test...")
result_n, trajectory_n = run_optimization_method("Modified Newton", "Extended Powell Singular", initial_point, tol=1e-5, max_iter=1000)
print_result(result_n)


print("\nRunning Steepest Descent Test...")
result_s, trajectory_s = run_optimization_method("Steepest Descent", "Extended Powell Singular", initial_point, tol=1e-5, max_iter=1000)
print_result(result_s)

plot_combined_results(extended_powell_singular, trajectory_n, trajectory_s, 'Extended Powell Singular', 'initial', [0, 0], [0, 0])
```

#### Random Points

```python
# Optimize for each random starting point
for idx, random_point in enumerate(random_points):
    print(f"\nCOST OF EXTENDED POWELL SINGULAR BEFORE OPTIMIZATION (Random Point {idx+1}):",
          extended_powell_singular(random_point))
    
    print(f"\nRunning Modified Newton Test for Random Point {idx+1}...")
    result_n, trajectory_n = run_optimization_method("Modified Newton", "Extended Powell Singular", random_point, tol=1e-5, max_iter=1000)
    print_result(result_n)

    print(f"\nRunning Steepest Descent Test for Random Point {idx+1}...")
    result_s, trajectory_s = run_optimization_method("Steepest Descent", "Extended Powell Singular", random_point, tol=1e-5, max_iter=1000)
    print_result(result_s)

    plot_combined_results(extended_powell_singular, trajectory_n, trajectory_s, 'Extended Powell Singular', f'random_{idx+1}')
```
