#### Dependencies

```python
import numpy as np
from algorithms.utils.extended_rosenbrock import extended_rosenbrock
from algorithms.utils.plot import plot_combined_results, plot
from algorithms.utils.print_result import print_result
from tests.test_optimization_method import run_optimization_method

np.random.seed(336648)
```

#### Initialization

```python
# Define the problem base and scale
base_number = 10  # Base number for scaling
dimension_power = 3  # Exponent for the problem size
problem_size = base_number ** dimension_power  # Total problem size

# Initialize the starting point based on x_bar
initial_point = np.zeros(problem_size)
for l in range(1, problem_size + 1):
    if l % 2 == 1:  # mod(l, 2) = 1
        initial_point[l - 1] = -1.2
    else:  # mod(l, 2) = 0
        initial_point[l - 1] = 1.0

# Generate 10 random starting points in the hypercube [x - 1, x + 1]
random_points = [initial_point + np.random.uniform(-1, 1, problem_size) for _ in range(10)]
```

#### Initial Points

```python
# Print the cost before optimization
print("\nCOST OF EXTENDED ROSENBROCK BEFORE OPTIMIZATION:", extended_rosenbrock(initial_point))

# Run Modified Newton Method
print("\nRunning Modified Newton Test...")
result_n, trajectory_n = run_optimization_method("Modified Newton", "Extended Rosenbrock", initial_point, tol=1e-5, max_iter=1000)
print_result(result_n)

# Run Steepest Descent Method
print("\nRunning Steepest Descent Test...")
result_s, trajectory_s = run_optimization_method("Steepest Descent", "Extended Rosenbrock", initial_point, tol=1e-5, max_iter=1000)
print_result(result_s)

# Plot the combined results
plot_combined_results(extended_rosenbrock, trajectory_n, trajectory_s, "Extended Rosenbrock", "initial", [-1, 1], [-1, 1])
```

#### Random points

```python
# Optimize for each random starting point
for idx, random_point in enumerate(random_points):
    print(f"\nCOST OF Extended Rosenbrock BEFORE OPTIMIZATION (Random Point {idx+1}):",
          extended_rosenbrock(random_point))
    
    print(f"\nRunning Modified Newton Test for Random Point {idx+1}...")
    result_n, trajectory_n = run_optimization_method("Modified Newton", "Extended Rosenbrock", random_point, tol=1e-5, max_iter=1000)
    print_result(result_n)

    print(f"\nRunning Steepest Descent Test for Random Point {idx+1}...")
    result_s, trajectory_s = run_optimization_method("Steepest Descent", "Extended Rosenbrock", random_point, tol=1e-5, max_iter=1000)
    print_result(result_s)

    plot_combined_results(extended_rosenbrock, trajectory_n, trajectory_s, "Extended Rosenbrock", f"random_{idx+1}")
```
